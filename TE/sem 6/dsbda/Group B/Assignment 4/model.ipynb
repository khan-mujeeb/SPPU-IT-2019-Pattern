Visualize the data using Python libraries matplotlib, seaborn by plotting the graphs for
Tips dataset.
( Charts : Line chart, Barplot, Heatmap, Scatterplot, histogram,
boxplot, violin, timeseries chart)

```python
import pandas as pd 
df = pd.read_csv(r"F:\codes\dsbda\Group B\PARTB-DATASETS\tips_DV.csv")
```

```python
df.head()
```

```python
import seaborn as sns
```

Line Plot

The lineplot function in Seaborn provides a convenient way to visualize trends and patterns over time or any other continuous variable. It is particularly useful for displaying data that changes continuously, such as stock prices, temperature variations, or trends in data over time.

```python
sns.lineplot(data = df)
```

Barplot

The barplot() function in seaborn allows you to visualize the relationship between a categorical variable and a numerical variable. It displays the average value of the numerical variable for each category as a bar, along with error bars representing the confidence interval.

```python
sns.barplot(data = df, x = 'day', y = 'tip', hue = 'sex')
```

Heatmap

heatmap is a popular type of plot that allows you to create visual representations of data using color gradients.

```python
temp = df = pd.get_dummies(data= df)
sns.heatmap(data = temp)
```

Scatterplot

```python
sns.scatterplot(data=df)
```

histogram

 It provides a convenient way to visualize the distribution of a single variable and examine its underlying characteristics, such as the shape, central tendency, and spread of the data.

```python
sns.histplot(data=df, x = 'total_bill')
```


boxplot

```python
sns.boxplot(data = df['tip'])
```

violin

a violin plot is a method used to visualize the distribution of numerical data across different categories. It is a combination of a box plot and a kernel density plot

```python
sns.violinplot(data=df)
```
