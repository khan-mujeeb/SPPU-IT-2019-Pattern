```python
import pandas as pd
```

```python
dataset = pd.read_csv(r"F:\codes\dsbda\facbook_dataset.csv", sep = ';')
dataset.head()
```

A. subset

```python
subset1 = dataset[['Page total likes', 'Type']].loc[:50]
subset2 = dataset[['Paid']].loc[:50]
```

B. Merge

```python
subset3 = pd.concat([subset1, subset2])

subset3
```

C. sort

```python
sortetd_data = subset3.sort_values(['Page total likes'], ascending=False)
sortetd_data
```

D. DataShape and Reshape

```python


reshaped_df = pd.pivot_table(dataset, index=['Type', 'Category'], values='Page total likes')
reshaped_df
```
