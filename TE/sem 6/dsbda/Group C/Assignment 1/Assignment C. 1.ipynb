```python
# importing required libraries
import requests
import bs4
```

```python
request1 = requests.get('https://www.flipkart.com/poco-c31-royal-blue-64-gb/p/itm19effae969b86')
request1
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

```python
reviews = soup.findAll('div', {'class': 't-ZTKy'})
reviews
```

```python
review_list = []
for review in reviews:
    print(review.get_text() + "\n\n")
    review_list.append(review.get_text())
```

```python
product_rating = soup.find('div', {'class': '_2d4LTz'} ).get_text()

print("product rating is ", product_rating)
```

```python
individual_rating = soup.findAll('div', {'class': '_3LWZlK _1BLPMq'})
```

```python
rating_list = []

for rating in individual_rating:
    rating_list.append(rating.get_text())
    print(rating.get_text())
```

```python
review_rating_list = dict(zip(review_list, rating_list))
rating_list
```

```python
cust_name = soup.findAll('p', {'class': '_2sc7ZR _2V5EHH' })
```

```python
cust_list = []
for cust in cust_name:
    cust_list.append(cust.get_text())
    print(cust.get_text() + "\n\n")
    
```

```python
for customer, rating, review in zip(cust_list, rating_list, review_list):
    print(f"Customer: {customer}\nRating: {rating}\nReview: {review}\n")
```
